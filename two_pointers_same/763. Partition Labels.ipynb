```python
s = "ababcbacadefegdehijhklij"
# Output: [9,7,8]
from collections import defaultdict
```

```python
def partitionLabels(s):
    last_idx = {}
    for i, c in enumerate(s):
        last_idx[c] = i
    
    res = []
    n = len(s)
    left = 0
    while left < n:
        right = last_idx[s[left]]
        start = left
        while left < right:
            right = max(right, last_idx[s[left]])
            left += 1
        res.append(right - start + 1)
        left += 1
    return res

partitionLabels(s) # time: O(n), space: O(1)
        
        
```
