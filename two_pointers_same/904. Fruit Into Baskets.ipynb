```python
fruits = [3,3,3,1,2,1,1,2,3,3,4]
# output: 5
from collections import defaultdict
```

```python
def totalFruit1(fruits):
    basket = defaultdict(int)
    right = 0
    n = len(fruits)
    res = 0
    for left in range(n):
        while right < n and len(basket) <= 2:
            basket[fruits[right]] += 1
            right += 1
            if len(basket) <= 2:
                res = max(res, right - left)
        basket[fruits[left]] -= 1
        if basket[fruits[left]] == 0:
            del basket[fruits[left]]
    return res

totalFruit1(fruits) # time: O(n), space: O(1)
```

```python
def totalFruit2(fruits):
    basket = defaultdict(int)
    left = 0
    res = 0
    for right, fruit in enumerate(fruits):
        basket[fruit] += 1
        if len(basket) > 2:
            basket[fruits[left]] -= 1
            if basket[fruits[left]] == 0:
                del basket[fruits[left]]
            left += 1
        res = max(res, right - left + 1)
    return res

totalFruit2(fruits) # time: O(n), space: O(1)
```
