```python
heights = [10,6,8,5,11,9]
# Output: [3,1,2,1,1,0]
```

```python
def canSeePersonsCount(heights):
    n = len(heights)
    ans = [0] * n
    stack = []
    for i in reversed(range(n)):
        while stack and heights[stack[-1]] <= heights[i]:
            stack.pop()
            ans[i] += 1
        if stack:
            ans[i] += 1
        stack.append(i)
    return ans

canSeePersonsCount(heights) # time O(n) space O(n)
```
