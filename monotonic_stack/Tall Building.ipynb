```python
arr = [5,3,8,3,2,5]
# Output:[3,3,5,4,4,4]
```

```python
def count_buildings(arr, res, indices):
    stack = []
    for i in indices:
        res[i] += len(stack)
        while stack and arr[stack[-1]] <= arr[i]:
            stack.pop()
        stack.append(i)
    return res

            
def tall_building(arr):
    n = len(arr)
    res = [1] * n
    res = count_buildings(arr, res, range(n))
    res = count_buildings(arr, res, reversed(range(n)))
    return res

tall_building(arr) # time O(n) space O(n)
            
```
