```python
prices = [8,4,6,2,3]
# Output: [4,2,4,2,3]
```

```python
def finalPrices(prices):
    n = len(prices)
    ans = list(prices)
    stack = []
    for i in range(n):
        while stack and prices[stack[-1]] >= prices[i]:
            ans[stack[-1]] -= prices[i]
            stack.pop()
        stack.append(i)
    return ans

finalPrices(prices) # time O(n) space O(n)
```

```python
def finalPrices(prices):
    n = len(prices)
    ans = list(prices)
    stack = []
    for i in reversed(range(n)):
        while stack and prices[stack[-1]] >= prices[i]:
            stack.pop()
        if stack:
            ans[i] -= prices[stack[-1]]
        stack.append(i)
    return ans

finalPrices(prices) # time O(n) space O(n)
```
