```python
heights = [2,1,5,6,2,3]
# Output: 10
```

```python
def largestRectangleArea(heights):
    n = len(heights)
    max_area = 0
    stack = []
    for i in range(n + 1):
        val = heights[i] if i < n else -1
        while stack and heights[stack[-1]] > val:
            height = heights[stack.pop()]
            left = stack[-1] if stack else -1
            max_area = max(max_area, height * (i - left - 1))
        stack.append(i)
    return max_area

largestRectangleArea(heights) # time O(n) space O(n)
```
