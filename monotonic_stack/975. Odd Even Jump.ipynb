```python
arr = [5,1,3,4,2]
# Output: 3
```

```python
from collections import deque
def OddEvenJump(arr):
    n = len(arr)
    graph = build_graph(arr)
    visited = set([(n - 1, 0), (n - 1, 1)])
    q = deque([(n - 1, 0), (n - 1, 1)])
    ans = 1
    
    while q:
        curr = q.popleft()
        for neighbor in graph[curr]:
            if neighbor not in visited:
                visited.add(neighbor)
                q.append(neighbor)
                if neighbor[1] ==1:
                    ans += 1
            
    return ans

def build_graph(arr):
    n = len(arr)
    graph = {}
    for i in range(n):
        graph[(i, 0)] = []
        graph[(i, 1)] = []

    indices = sorted(range(n), key=lambda i: arr[i])
    stack = []
    for i in indices:
        while stack and stack[-1] < i:
            j = stack.pop()
            graph[(i, 0)].append((j, 1))
        stack.append(i)

    indices = sorted(range(n), key=lambda i: -arr[i])
    stack = []
    for i in indices:
        while stack and stack[-1] < i:
            j = stack.pop()
            graph[(i, 1)].append((j, 0))
        stack.append(i)
    
    return graph
                
        
OddEvenJump(arr) # time O(nlogn) space O(n)
```
