```python
nums = [1,3,-1,-3,5,3,6,7]
k = 3
# Output: [3,3,5,5,6,7]
```

```python
from collections import deque

def maxSlidingWindow(nums, k):
    n = len(nums)
    q = deque()
    res = []
    for i in range(n):
        while q and nums[q[-1]] <= nums[i]:
            q.pop()
        q.append(i)
        if i < k - 1:
            continue
        res.append(nums[q[0]])
        if q[0] == i - k + 1:
            q.popleft()
    return res
    
maxSlidingWindow(nums, k) # time O(n) space O(k)
```
