```python
from collections import defaultdict
class TrieNode:
    def __init__(self):
        self.children = {}
        self.words = []

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
            node.words.append(word)

    def get_words(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return []
            node = node.children[char]
        return node.words
        
        
```

```python
def word_squares(words):
    trie = Trie()
    for word in words:
        trie.insert(word)

    res = []
    for word in words:
        dfs(trie, res, words, [word])
    return res

def dfs(trie, res, words, path):
    curr_idx = len(path)    
    if curr_idx == len(words[0]):
        res.append(path[:])
        return

    prefix = ''.join([path[i][curr_idx] for i in range(curr_idx)])
    candidates = trie.get_words(prefix)
    for word in candidates:
        path.append(word)
        dfs(trie, res, words, path)
        path.pop()

word_squares(words)
```
