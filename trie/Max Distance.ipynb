```python
S = ["011000","0111010","01101010"]
# 9
```

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False
        self.height = 0

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for i, char in enumerate(word):
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
            # 0 1 2
            # a b c
            node.height = max(node.height, len(word) - i - 1)
        node.is_end = True
        
```

```python
def maxDistance(S):
    trie = Trie()
    for word in S:
        trie.insert(word)
    return dfs(trie.root)

def dfs(node):
    result = 0
    if len(node.children) == 2:
        result = max(result, node.children["0"].height + node.children["1"].height + 2)

    if node.is_end:
        result = max(result, node.height)

    for child in node.children.values():
        result = max(result, dfs(child))
    return result

print(maxDistance(S)) # time O(n * L), space O(n * L)
```
