```python
nums = [1, 2, 3, 4]
# output: 7
```

```python
class TrieNode:
    def __init__(self):
        self.sons = [None, None]
        self.val = None

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, num):
        node = self.root
        for i in range(30, -1, -1):
            bit = (num >> i) & 1
            if not node.sons[bit]:
                node.sons[bit] = TrieNode()
            node = node.sons[bit]
        node.val = num

    def find_max_xor(self, num):
        node = self.root
        for i in range(30, -1, -1):
            bit = (num >> i) & 1
            if node.sons[1 - bit]:
                node = node.sons[1 - bit]
            else:
                node = node.sons[bit]
        return node.val
```

```python
def maximumXOR(nums):
    trie = Trie()
    max_xor = 0
    prefix_xor = 0
    trie.insert(prefix_xor)
    for num in nums:
        prefix_xor ^= num
        trie.insert(prefix_xor)
        candidate = trie.find_max_xor(prefix_xor)
        max_xor = max(max_xor, candidate ^ prefix_xor)
    return max_xor
```

```python
maximumXOR(nums) # time O(32n) space O(32n)
```
