```python
words = ["abc", "adc", "abd", "abcd"]
target = "ac"
k = 1
```

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False
        self.word = None
class Trie:
    def __init__(self):
        self.root = TrieNode() 
    
    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True
        node.word = word
        
```

```python
def kEditDistance(words, target, k):
    n = len(target)
    trie = Trie()
    for word in words:
        trie.insert(word)
    result = []
    dp = list(range(n + 1))
    return dfs(trie.root, target, k, dp, result)

def dfs(node, target, k, last_dp, result):
    n = len(target)
    if node.is_end and last_dp[n] <= k:
        result.append(node.word)
    
    for char in node.children:
        current_dp = list(range(n + 1))
        current_dp[0] = last_dp[0] + 1
        for i in range(1, n + 1):
            # delete
            current_dp[i] = last_dp[i] + 1
            # insert
            current_dp[i] = min(current_dp[i], current_dp[i - 1] + 1)
            # replace
            if target[i - 1] == char:
                current_dp[i] = min(current_dp[i], last_dp[i - 1])
            else:
                current_dp[i] = min(current_dp[i], last_dp[i - 1] + 1)
        dfs(node.children[char], target, k, current_dp, result)
    return result

kEditDistance(words, target, k) # time: O(sizeOfTrie * n) space: O(sizeOfTrie + heightOfTrie * n)
    
```
