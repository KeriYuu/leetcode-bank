```python
nums = [3,10,5,25,2,8]
# Output: 28
```

```python
class TrieNode:
    def __init__(self):
        self.sons = [None, None]
        self.val = None

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, num):
        node = self.root
        for i in range(30, -1, -1):
            bit = (num >> i) & 1
            if not node.sons[bit]:
                node.sons[bit] = TrieNode()
            node = node.sons[bit]
        node.val = num

    def find_max_xor(self, num):
        node = self.root
        for i in range(30, -1, -1):
            bit = (num >> i) & 1
            if node.sons[1 - bit]:
                node = node.sons[1 - bit]
            else:
                node = node.sons[bit]
        return node.val
```

```python
def maximumXOR(nums):
    trie = Trie()
    max_xor = 0

    for num in nums:
        trie.insert(num)
        candidate = trie.find_max_xor(num)
        max_xor = max(max_xor, candidate ^ num)
    return max_xor
```

```python
maximumXOR(nums) # time O(32n) space O(32n)
```
