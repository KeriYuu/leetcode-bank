```python
matrix = ["doaf", "agai", "dcan"]
words = ["dog", "dad", "dgdg", "can", "again"]
# output: 2
```

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True
    
```

```python
def wordSearchIII(matrix, words):
    trie = Trie()
    for word in words:
        trie.insert(word)
    res = [0]
    word_count = 0
    m, n = len(matrix), len(matrix[0])
    visited = [[False] * n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            if visited[i][j]:
                continue
            dfs(visited, matrix, trie.root, res, trie, word_count, i, j, i, j)
    return res[0]

def dfs(visited, matrix, root, res, trie, word_count, i, j, start_x, start_y):
    m, n = len(matrix), len(matrix[0])
    if  matrix[i][j] not in root.children:
        return
    
    node = root.children[matrix[i][j]]
    visited[i][j] = True
    if node.is_end:
        word_count += 1
        res[0] = max(res[0], word_count)
        for ni in range(start_x, m):
            _start_y = start_y if ni == start_x else 0
            for nj in range(_start_y, n):
                if visited[ni][nj]:
                    continue
                dfs(visited, matrix, trie.root, res, trie, word_count, ni, nj, ni, nj)
        word_count -= 1


    for di, dj in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
        ni, nj = i + di, j + dj
        if 0 <= ni < m and 0 <= nj < n and not visited[ni][nj]:
            dfs(visited, matrix, node, res, trie, word_count, ni, nj, start_x, start_y)

    visited[i][j] = False
    
```

```python
wordSearchIII(matrix, words)
```
