```python
length = 15
connections = [[2, 8], [6, 9]]
# Output: 2
```

```python
from collections import deque

def build_graph(length, connections):
    graph = {i : set() for i in range(1, length+1)}
    for a, b in connections:
        graph[a].add(b)
    return graph

def modern_ludo_spfa(length, connections):
    graph = build_graph(length, connections)

    q = deque([1])
    dist = {i: float('inf') for i in range(1, length+1)}
    dist[1] = 0

    while q:
        current_node = q.popleft()
        for direct_node in graph[current_node]:
            if dist[direct_node] > dist[current_node]:
                dist[direct_node] = dist[current_node]
                q.append(direct_node)
        for next_node in range(current_node+1, min(current_node+7, length+1)):
            if dist[next_node] > dist[current_node] + 1:
                dist[next_node] = dist[current_node] + 1
                q.append(next_node)

    return dist[length]

modern_ludo_spfa(length, connections) # time O(n) -> O(n * m), space O(n + m), n: length, m: len(connections)
```

```python
import heapq
def modern_ludo_dijkstra(length, connections):
    graph = build_graph(length, connections)

    pq = [(0, 1)]
    dist = {i: float('inf') for i in range(1, length+1)}
    dist[1] = 0

    while pq:
        current_dist, current_node = heapq.heappop(pq)
        for next_node in graph[current_node]:
            if dist[next_node] > dist[current_node]:
                dist[next_node] = dist[current_node]
                heapq.heappush(pq, (dist[next_node], next_node))

        for next_node in range(current_node+1, min(current_node+7, length+1)):
            if dist[next_node] > dist[current_node] + 1:
                dist[next_node] = dist[current_node] + 1
                heapq.heappush(pq, (dist[next_node], next_node))

    return dist[length]

modern_ludo_dijkstra(length, connections) # time O((n + m) log(n)), space O(n + m), n: length, m: len(connections)
```
