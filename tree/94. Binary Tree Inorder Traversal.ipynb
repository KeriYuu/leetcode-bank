```python
tree = [1,None,2,3]

# Output: [1,3,2]
```

```python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def buildTree(arr):
    if not arr:
        return None
    root = TreeNode(arr[0])
    queue = deque([root])
    n = len(arr)
    i = 1
    while queue:
        node = queue.popleft()
        if i < n and arr[i] is not None:
            node.left = TreeNode(arr[i])
            queue.append(node.left)
        i += 1
        if i < n and arr[i] is not None:
            node.right = TreeNode(arr[i])
            queue.append(node.right)
        i += 1
    return root

def serialize(root):
    if not root:
        return []
    queue = deque([root])
    result = []
    while queue:
        node = queue.popleft()
        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)
    while result and result[-1] is None:
        result.pop()
    return result

root = buildTree(tree)
print(serialize(root))

```

```python
def inorderTraversal(root):
    res, stack = [], []
    cur = root
    while cur or stack:
        while cur:
            stack.append(cur)
            cur = cur.left
        cur = stack.pop()
        res.append(cur.val)
        cur = cur.right
    return res

```

```python
inorderTraversal(root) # time: O(n), space: O(n)
```

```python
def inorderTraversal2(root):
    res, stack = [], []
    while root is not None:
        stack.append(root)
        root = root.left
    while stack:
        node = stack[-1]
        res.append(node.val)
        if node.right is not None:
            root = node.right
            while root is not None:
                stack.append(root)
                root = root.left
        else:
            stack.pop()
            while stack and stack[-1].right == node:
                node = stack.pop()
    return res

inorderTraversal2(root) # time: O(n), space: O(n)
```

```python
def inorderTraversal_recursive(root):
    def helper(node, result):
        if node:
            helper(node.left, result)
            result.append(node.val)
            helper(node.right, result)
    result = []
    helper(root, result)
    return result
```

```python
inorderTraversal_recursive(root) # time: O(n), space: O(n)
```
