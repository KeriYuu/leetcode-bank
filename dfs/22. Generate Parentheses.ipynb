```python
n = 3
# Output: ["((()))","(()())","(())()","()(())","()()()"]
```

```python
def generateParenthesis(n):
    res = []
    dfs(0, n, [], res)
    return res

def dfs(num_left, n, path, res):
    if len(path) == 2 * n:
        res.append(''.join(path))
        return
    if num_left < n:
        path.append('(')
        dfs(num_left + 1, n, path, res)
        path.pop()
    if 2 * num_left > len(path):
        path.append(')')
        dfs(num_left, n, path, res)
        path.pop()
```

```python
generateParenthesis(n) # time: O(n * C_n), space: O(n * C_n)

# C_n is the nth Catalan number
# The number of distinct full binary tree structures with n internal nodes (or equivalently, with n+1 leaves) is the nth Catalan number.
```
