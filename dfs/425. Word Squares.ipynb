```python
words = ["area","lead","wall","lady","ball"]
# Output:
# [["wall","area","lead","lady"],["ball","area","lead","lady"]]
```

```python
# brute force: C_n_l * l! = A_n_l
```

```python
def word_squares(words):
    prefix_dict = build_prefix_dict(words)
    res = []
    for word in words:
        dfs(res, prefix_dict, words, [word])
    return res

def build_prefix_dict(words):
    prefix_dict = {}
    for word in words:
        for i in range(len(word)):
            prefix = word[:i+1]
            if prefix not in prefix_dict:
                prefix_dict[prefix] = []
            prefix_dict[prefix].append(word)
    return prefix_dict

def dfs(res, prefix_dict, words, path):
    curr_idx = len(path)
    if curr_idx == len(words[0]):
        res.append(path[:])
        return
    
    # pruning:
    for prune_idx in range(curr_idx, len(words[0])):
        prefix = ''.join([path[i][prune_idx] for i in range(curr_idx)])
        if prefix not in prefix_dict:
            return

    prefix = ''.join([path[i][curr_idx] for i in range(curr_idx)])
    
    for word in prefix_dict[prefix]:
        path.append(word)
        dfs(res, prefix_dict, words, path)
        path.pop()

word_squares(words)
```
