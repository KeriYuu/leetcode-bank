```python
arr = [5,1,3,4,2]
# Output: 3
```

```python
def OddEvenJump(arr):
    even_jumps = build_even_jumps(arr)
    odd_jumps = build_odd_jumps(arr)
    n = len(arr)
    ans = 0
    for i in range(n):
        j = i
        next_jumps = odd_jumps
        while j < n:
            if next_jumps[j] is None:
                break
            j = next_jumps[j]
            next_jumps = even_jumps if next_jumps == odd_jumps else odd_jumps
        if j == n - 1:
            ans += 1
    return ans
    
def build_odd_jumps(arr):
    n = len(arr)
    jumps = [None] * n
    for i in range(n):
        min_val = float('inf')
        min_idx = -1
        for j in range(i + 1, n):
            if arr[j] < arr[i]:
                continue
            if arr[j] < min_val:
                min_val = arr[j]
                min_idx = j

        if min_idx != -1:
            jumps[i] = min_idx
    return jumps
    
def build_even_jumps(arr):
    n = len(arr)
    jumps = [None] * n
    for i in range(n):
        max_val = float('-inf')
        max_idx = -1
        for j in range(i + 1, n):
            if arr[j] > arr[i]:
                continue
            if arr[j] > max_val:
                max_val = arr[j]
                max_idx = j

        if max_idx != -1:
            jumps[i] = max_idx
    return jumps
    
    
```

```python
OddEvenJump(arr) # time O(n^2 + n^2 + n^2), space O(n)
```
