```python
n = 3
k = 2
row = 0
column = 0
# Output: 0.06250
```

```python
def knightProbability(n, k, row, column):
    return memo_search(n, k, row, column, {})

def memo_search(n, k, row, column, memo):
    if row < 0 or row >= n or column < 0 or column >= n:
        return 0.0

    if k == 0:
        memo[(k, row, column)] = 1.0
        return 1.0

    if (k, row, column) in memo:
        return memo[(k, row, column)]
    
    prob = 0.0
    for dx, dy in [(-2, -1), (-2, 1), (-1, -2), (-1, 2), (1, -2), (1, 2), (2, -1), (2, 1)]:
        nx, ny = row + dx, column + dy
        prob += memo_search(n, k - 1, nx, ny, memo) / 8.0

    memo[(k, row, column)] = prob
    return prob

knightProbability(n, k, row, column) # time O(n^2 * k), space O(k * n^2)
```
