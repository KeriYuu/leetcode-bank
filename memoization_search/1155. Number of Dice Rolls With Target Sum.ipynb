```python
n = 1
k = 6
target = 3
# Output: 1
```

```python
def numRollsToTarget(n, k, target):
    memo = {}
    return dfs(n, k, target, memo)

def dfs(n, k, target, memo):
    MOD = 10**9 + 7
    
    if n == 0:
        return 1 if target == 0 else 0

    if target < n or target > n * k:
        return 0

    if (n, target) in memo:
        return memo[(n, target)]
    
    res = 0
    for face in range(1, k + 1):
        if target - face >= 0:
            res += dfs(n - 1, k, target - face, memo)
            res %= MOD

    memo[(n, target)] = res
    return res

numRollsToTarget(n, k, target) # time O(n * k * target) space O(n * target)
```
