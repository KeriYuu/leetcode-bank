```python
height = [0,1,0,2,1,0,1,3,2,1,2,1]
# Output: 6
```

```python
# prefix array
def trap(height):
    n = len(height)

    left_max = [0] * n
    right_max = [0] * n

    left_max[0] = height[0]
    for i in range(1, n):
        left_max[i] = max(left_max[i - 1], height[i])
        
    right_max[n - 1] = height[n - 1]
    for i in range(n - 2, -1, -1):
        right_max[i] = max(right_max[i + 1], height[i])

    trapped_water = 0
    for i in range(n):
        trapped_water += min(left_max[i], right_max[i]) - height[i]

    return trapped_water

trap(height) # time O(n), space O(n)


```

```python
# two pointers
def trap(height):
    n = len(height)
    left, right = 0, n - 1
    trapped_water = 0

    while left < right:
        if height[left] < height[right]:
            if height[left] > height[left + 1]:
                trapped_water += height[left] - height[left + 1]
                height[left + 1] = height[left]
            left += 1
        else:
            if height[right] > height[right - 1]:
                trapped_water += height[right] - height[right - 1]
                height[right - 1] = height[right]
            right -= 1
    return trapped_water

trap(height) # time O(n), space O(1)
    
```
