```python
n = 3
k = 2
row = 0
column = 0
# Output: 0.06250
```

```python
def knightProbability(n, k, row, column):
    prev = [[0.0] * n for _ in range(n)]
    prev[row][column] = 1.0
    for _ in range(k):
        curr = [[0] * n for _ in range(n)]
        for r in range(n):
            for c in range(n):
                if prev[r][c] == 0.0:
                    continue
                for dr, dc in [(-2, -1), (-2, 1), (-1, -2), (-1, 2), (1, -2), (1, 2), (2, -1), (2, 1)]:
                    nr, nc = r + dr, c + dc
                    if nr < 0 or nr >= n or nc < 0 or nc >= n:
                        continue
                    curr[nr][nc] += prev[r][c] / 8.0
        prev = curr

    return sum(map(sum, prev))
                    

knightProbability(n, k, row, column) # time O(n^2 * k), space O(n^2)
```
