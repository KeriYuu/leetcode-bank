```python
steps = 3
arrLen = 2
# Output: 4
```

```python
def numWays(steps, arrLen):
    n = min(steps // 2 + 1, arrLen)
    if n <= 1:
        return 1
    dp = [[0] * n for _ in range(2)]
    dp[0][0] = 1

    for step in range(1, steps + 1):
        dp[step % 2][0] = dp[(step - 1) % 2][0] + dp[(step - 1) % 2][1]
        dp[step % 2][n - 1] = dp[(step - 1) % 2][n - 1] + dp[(step - 1) % 2][n - 2]
        for i in range(1, n - 1):
            dp[step % 2][i] = dp[(step - 1) % 2][i - 1] + dp[(step - 1) % 2][i] + dp[(step - 1) % 2][i + 1]

    return dp[steps % 2][0]

numWays(steps, arrLen) # time O(steps * n), space O(n)
            
```
