```python
s = "abc"
# Output: 7
```

```python
def distinctSubseqII(s):
    n = len(s)
    visited = set()
    dp = [0] * n
    for i in range(n):
        if s[i] not in visited:
            visited.add(s[i])
            dp[i] = 1

    for i in range(n):
        for j in range(i - 1, -1, -1):
            dp[i] += dp[j]
            if s[j] == s[i]:
                break

    return sum(dp)

distinctSubseqII(s) # time O(n**2), space O(n)
```
