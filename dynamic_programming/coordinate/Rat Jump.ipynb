```python
def RatJump(arr):
    n = len(arr)
    dp = [[0, 0] for _ in range(5)]

    even_jump = [1, 3, 4]
    odd_jump = [1, 2, 4]

    dp[0][0] = 1

    for i in range(1, n - 1):
        dp[i % 5] = [0, 0] # reset dp[i % 5]!
        if arr[i % 5] == 1:
            continue
        for jump in even_jump:
            if i - jump >= 0:
                dp[i % 5][0] = dp[(i - jump) % 5][1]
        for jump in odd_jump:
            if i - jump >= 0:
                dp[i % 5][1] = dp[(i - jump) % 5][0]

    total = 0
    for jump in even_jump:
        for i in range(max(0, n - jump - 1), n - 1):
            total += dp[i % 5][1]
    for jump in odd_jump:
        for i in range(max(0, n - jump - 1), n - 1):
            total += dp[i % 5][0]
    return total

RatJump(arr) # time O(n), space O(n)
    
```
