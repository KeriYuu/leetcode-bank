```python
n = 1
k = 6
target = 3
# Output: 1
```

```python
def numRollsToTarget(n, k, target):
    MOD = 10**9 + 7
    prev = [0] * (target + 1)
    prev[0] = 1
    for _ in range(n):
        curr = [0] * (target + 1)
        for s in range(1, target + 1):
            for face in range(1, k + 1):
                if s - face >= 0:
                    curr[s] = (curr[s] + prev[s - face]) % MOD
        prev = curr
    return prev[target]

numRollsToTarget(n, k, target)
```

```python
# time O(n * k * target) space O(target)
```
