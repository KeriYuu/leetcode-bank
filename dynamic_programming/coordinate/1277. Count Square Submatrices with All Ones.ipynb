```python
matrix = [
  [0,1,1,1],
  [1,1,1,1],
  [0,1,1,1]
]
# Output: 15
```

```python
def countSquares(matrix):
    m, n = len(matrix), len(matrix[0])

    dp = [[0] * n for _ in range(m)]
    up = [[0] * n for _ in range(m)]
    left = [[0] * n for _ in range(m)]

    for i in range(m):
        dp[i][0] = left[i][0] = up[i][0] = matrix[i][0]

    for j in range(n):
        dp[0][j] = left[0][j] = up[0][j] = matrix[0][j]

    for i in range(1, m):
        for j in range(1, n):
            if matrix[i][j] == 0:
                continue
            up[i][j] = up[i-1][j] + 1
            left[i][j] = left[i][j-1] + 1
            dp[i][j] = min(up[i][j], left[i][j], dp[i-1][j-1] + 1)

    total = 0
    for i in range(m):
        for j in range(n):
            total += dp[i][j]
    return total

countSquares(matrix) # time O(m*n), space O(m*n)
```

```python
def countSquares_rolling_array(matrix):
    m, n = len(matrix), len(matrix[0])
    dp = [[0] * n for _ in range(2)]
    up = [[0] * n for _ in range(2)]
    left = [[0] * n for _ in range(2)]
    
    for j in range(n):
        dp[0][j] = left[0][j] = up[0][j] = matrix[0][j]

    answer = sum(dp[0])

    for i in range(1, m):
        dp[i % 2][0] = left[i % 2][0] = up[i % 2][0] = matrix[i][0]
        for j in range(1, n):
            if matrix[i][j] == 0:
                dp[i % 2][j] = left[i % 2][j] = up[i % 2][j] = 0
                continue
            up[i % 2][j] = up[(i - 1) % 2][j] + 1
            left[i % 2][j] = left[i % 2][j-1] + 1
            dp[i % 2][j] = min(up[i % 2][j], left[i % 2][j], dp[(i - 1) % 2][j - 1] + 1)
        answer += sum(dp[i % 2])
    return answer

countSquares_rolling_array(matrix) # time O(m*n), space O(n)
```
