```python
p_time=[20,50,100,5]
p_score=[20,30,60,3]
f_time=[100,80,110,10]
f_score=[60,55,88,6]
# output: 94
```

```python
def exam_strategy(p_time, p_score, f_time, f_score):
    n = len(p_time)
    dp = [[0] * (120 + 1) for _ in range(2)]
    for i in range(n + 1):
        for j in range(120 + 1):
            dp[i % 2][j] = dp[(i - 1) % 2][j]
            if j >= p_time[i - 1]:
                dp[i % 2][j] = max(dp[i % 2][j], dp[(i - 1) % 2][j - p_time[i - 1]] + p_score[i - 1])
            if j >= f_time[i - 1]:
                dp[i % 2][j] = max(dp[i % 2][j], dp[(i - 1) % 2][j - f_time[i - 1]] + f_score[i - 1])
    return dp[n % 2][120]

exam_strategy(p_time, p_score, f_time, f_score) # time: O(n * t), space: O(t)
```
