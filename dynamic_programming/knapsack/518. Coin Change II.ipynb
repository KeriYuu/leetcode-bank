```python
amount = 5
coins = [1,2,5]
# Output: 4
```

```python
def change1(amount, coins):
    n = len(coins)
    dp = [[0] * (amount + 1) for _ in range(n + 1)]
    for i in range(n + 1):
        dp[i][0] = 1

    for i in range(1, n + 1):
        for j in range(1, amount + 1):
            dp[i][j] = dp[i - 1][j]
            if j >= coins[i - 1]:
                dp[i][j] += dp[i][j - coins[i - 1]]
    return dp[-1][-1]

change1(amount, coins) # time O(n * amount), space O(n * amount)
```

```python
def change2(amount, coins):
    n = len(coins)
    dp = [0] * (amount + 1)
    dp[0] = 1

    for i in range(n):
        for j in range(coins[i], amount + 1):
            dp[j] += dp[j - coins[i]]
    return dp[-1]

change2(amount, coins) # time O(n * amount), space O(amount)
```
