```python
length = 15
connections = [[2, 8], [6, 9]]
# Output: 2
```

```python
def build_graph(length, connections):
    graph = {i : set() for i in range(1, length+1)}
    for a, b in connections:
        graph[a].add(b)
    return graph

def modern_ludo(length, connections):
    graph = build_graph(length, connections)
    dist = [float('inf')] * (length + 1)
    dist[length] = 0
    for i in range(length - 1, 0, -1):
        for j in range(i + 1, min(i + 7, length + 1)):
            dist[i] = min(dist[i], dist[j] + 1)
        for j in graph[i]:
            dist[i] = min(dist[i], dist[j])
    return dist[1]

modern_ludo(length, connections) # time O(n + m), space O(n), n: length, m: len(connections)
```
