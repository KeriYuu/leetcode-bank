```python
arr = [5,1,3,4,2]
# Output: 3
```

```python
def OddEvenJump(arr):
    n = len(arr)
    graph = build_graph(arr)
    dp = [[0] * 2 for _ in range(n)]
    dp[n - 1][0] = 1
    dp[n - 1][1] = 1
    
    for i in range(n - 2, -1, -1):
        if (i, 0) in graph:
            dp[i][0] = dp[graph[(i, 0)]][1]
        if (i, 1) in graph:
            dp[i][1] = dp[graph[(i, 1)]][0]
    
    ans = 0
    for i in range(n):
        ans += dp[i][1]
    return ans

def build_graph(arr):
    n = len(arr)
    graph = {}

    indices = sorted(range(n), key=lambda i: arr[i])
    stack = []
    for i in indices:
        while stack and stack[-1] < i:
            j = stack.pop()
            graph[(j, 0)] = i
        stack.append(i)

    indices = sorted(range(n), key=lambda i: -arr[i])
    stack = []
    for i in indices:
        while stack and stack[-1] < i:
            j = stack.pop()
            graph[(j, 1)] = i
        stack.append(i)  

    return graph

OddEvenJump(arr) # time O(nlogn) space O(n)

        
    
```
