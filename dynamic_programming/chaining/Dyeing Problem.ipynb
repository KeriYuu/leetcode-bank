```python
n = 4
m = 3
# Output: 18
```

```python
def dyeing_problem(n, m):
    dp = [0] * (n + 1)
    dp[1] = m
    dp[2] = m * (m - 1)
    dp[3] = m * (m - 1) * (m - 2)
    for i in range(4, n + 1):
        dp[i] = dp[i - 1] * (m - 2) + dp[i - 2] * (m - 1)
    return dp[n]

dyeing_problem(n, m) # time O(n), space O(n)
```
