```python
S = "abcdebdde"
T = "bde"
# "bcde"
```

```python
def min_window(S, T):
    m, n = len(S), len(T)
    dp = [[float('inf')] * n for _ in range(m)]

    min_len = float('inf')
    res = ""

    for i in range(m):
        for j in range(n):
            if S[i] == T[j]:
                if j == 0:
                    dp[i][j] = i
                else:
                    dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = dp[i - 1][j]

        current_len = i - dp[i][n - 1] + 1
        if dp[i][n - 1] != float('inf'):
            if current_len < min_len:
                min_len = current_len
                res = S[dp[i][n - 1]:i + 1]

    return res

min_window(S, T) # time O(m * n), space O(m * n)
```
