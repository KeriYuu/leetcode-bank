```python
nums = [1,3,-1,-3,5,3,6,7]
k = 3
# Output: [3,3,5,5,6,7]
```

```python
import heapq
class Heap:
    def __init__(self):
        self.min_heap = []
        self.to_delete = set()
        
    def push(self, num, index):
        heapq.heappush(self.min_heap, (- num, index))
        
    def top(self):
        self._lazy_delete()
        return -self.min_heap[0][0]

    def delete(self, index):
        self.to_delete.add(index)
    
    def _lazy_delete(self):
        while self.min_heap and self.min_heap[0] in self.to_delete:
            heapq.heappop(self.min_heap)
        
        
```

```python
def maxSlidingWindow(nums, k):
    n = len(nums)
    heap = Heap()
    res = []
    for i in range(n):
        heap.push(nums[i], i)
        if i < k - 1:
            continue
        res.append(heap.top())
        heap.delete(i - k + 1)
    return res
```

```python
maxSlidingWindow(nums, k) # time O(nlogk) space O(k)
```
