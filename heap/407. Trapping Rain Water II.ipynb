```python
heightMap = [[1,4,3,1,3,2],[3,2,1,3,2,4],[2,3,3,2,3,1]]
# Output: 4
```

```python
from heapq import heappush, heappop

def trapRainWater(heightMap):
    m, n = len(heightMap), len(heightMap[0])

    heap = []
    visited = [[False] * n for _ in range(m)]

    for i in range(m):
        heappush(heap, (heightMap[i][0], i, 0))
        heappush(heap, (heightMap[i][n-1], i, n-1))
        visited[i][0] = True
        visited[i][n-1] = True

    for j in range(n):
        heappush(heap, (heightMap[0][j], 0, j))
        heappush(heap, (heightMap[m-1][j], m-1, j))
        visited[0][j] = True
        visited[m-1][j] = True

    directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
    res = 0

    while heap:
        height, x, y = heappop(heap)
        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if nx < 0 or nx >= m or ny < 0 or ny >= n or visited[nx][ny]:
                continue
            visited[nx][ny] = True
            if height > heightMap[nx][ny]:
                res += height - heightMap[nx][ny]
            heappush(heap, (max(height, heightMap[nx][ny]), nx, ny))
    return res
    
        
```

```python
trapRainWater(heightMap) # time O(mnlog(mn)) space O(mn)
```
