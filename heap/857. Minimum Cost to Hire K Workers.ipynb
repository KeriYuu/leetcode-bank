```python
quality = [10,20,5]
wage = [70,50,30]
k = 2
# Output: 105.00000
```

```python
from heapq import heappush, heappop

def mincostToHireWorkers(quality, wage, k):
    wage = [float(w) for w in wage]
    workers = sorted([(w / q, q) for q, w in zip(quality, wage)])

    heap = []
    res = float('inf')
    total_quality = 0

    for ratio, quality in workers:
        if len(heap) == k:
            total_quality -= -heappop(heap)
        heappush(heap, -quality)
        total_quality += quality
        if len(heap) == k:
            res = min(res, ratio * total_quality)
    return res
    
```

```python
mincostToHireWorkers(quality, wage, k) # time O(nlogn + nlogk) space O(n + k)
```
