```python
nums = [1,2,1,3,5,6,4]
# Output: 5
```

```python
def findPeakElement(nums):
    left, right = 0, len(nums) - 1
    while left + 1< right:
        mid = (left + right) // 2
        if nums[mid] < nums[mid + 1]:
            left = mid
        else:
            right = mid
    return left if nums[left] > nums[right] else right

findPeakElement(nums) # time: O(logn), space: O(1)
```
