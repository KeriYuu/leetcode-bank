```python
mat = [[10,20,15],[21,30,14],[7,16,32]]
# Output: [1,1]
```

```python
def findMaxCol(row):
    return row.index(max(row))

def findPeakGrid(mat):
    m = len(mat)
    down, up = 0, m - 1
    while down + 1 < up:
        mid = (down + up) // 2
        col = findMaxCol(mat[mid])
        if mid > 0 and mat[mid][col] < mat[mid - 1][col]:
            up = mid
        elif mid < m - 1 and mat[mid][col] < mat[mid + 1][col]:
            down = mid
        else:
            return [mid, col]

    col = findMaxCol(mat[down])
    if mat[down][col] > mat[down + 1][col] and (down == 0 or mat[down][col] > mat[down - 1][col]):
        return [down, col]
    col = findMaxCol(mat[up])
    return [up, col]

findPeakGrid(mat) # time: O(nlogm), space: O(1)
            
```
