```python
s = "abcde"
words = ["a","bb","acd","ace"]
# Output: 3
```

```python
from collections import defaultdict
def find_first_greater(pos, prev):
    left, right = 0, len(pos) - 1
    while left + 1 < right:
        mid = (left + right) // 2
        if pos[mid] <= prev:
            left = mid
        else:
            right = mid
    if pos[left] > prev:
        return left
    elif pos[right] > prev:
        return right
    else:
        return -1

def numMatchingSubseq(s, words):
    char2pos = defaultdict(list)
    for i, c in enumerate(s):
        char2pos[c].append(i)

    ans = 0
    for w in words:
        prev = -1
        flag = True
        for c in w:
            if not char2pos[c]:
                flag = False
                break
            pos = char2pos[c]
            index = find_first_greater(pos, prev)
            if index == -1:
                flag = False
                break
            prev = pos[index]
        if flag:
            ans += 1
    return ans
            
```

```python
numMatchingSubseq(s, words) # time: O(l * log(n)), space: O(n)
```
