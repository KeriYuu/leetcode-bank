```python
arr = [5,1,3,4,2]
# Output: 3
```

```python
from collections import deque
def OddEvenJump(arr):
    n = len(arr)
    graph = build_graph(arr)
    q = deque([(n - 1, 0), (n - 1, 1)])
    visited = set([(n - 1, 0), (n - 1, 1)])
    ans = 1
    while q:
        curr = q.popleft()
        for neighbor in graph[curr]:
            if neighbor in visited:
                continue
            visited.add(neighbor)
            q.append(neighbor)
            if neighbor[1] == 1:
                ans += 1

    return ans

def build_graph(arr):
    graph = {}
    n = len(arr)
    for i in range(n):
        graph[(i, 0)] = []
        graph[(i, 1)] = []

    for i in range(n):
        min_val = float('inf')
        min_idx = -1
        for j in range(i + 1, n):
            if arr[j] < arr[i]:
                continue
            if arr[j] < min_val:
                min_val = arr[j]
                min_idx = j

        if min_idx != -1:
            graph[(min_idx, 0)].append((i, 1))

    for i in range(n):
        max_val = float('-inf')
        max_idx = -1
        for j in range(i + 1, n):
            if arr[j] > arr[i]:
                continue
            if arr[j] > max_val:
                max_val = arr[j]
                max_idx = j

        if max_idx != -1:
            graph[(max_idx, 1)].append((i, 0))

    return graph
    
```

```python
OddEvenJump(arr) # time O(n^2) space O(n)
```
