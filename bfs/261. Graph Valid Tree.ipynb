```python
n = 5 
edges = [[0, 1], [0, 2], [0, 3], [1, 4]]
# Output: true.
```

```python
from collections import defaultdict, deque
def valid_tree(n, edges):
    if n - 1 != len(edges):
        return False
    
    graph = defaultdict(list)
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
        
    visited = set([0])
    q = deque([0])
    
    while q:
        node = q.popleft()
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                q.append(neighbor)
                
    return len(visited) == n

valid_tree(n, edges) # time: O(n + m), space: O(n + m)
```
