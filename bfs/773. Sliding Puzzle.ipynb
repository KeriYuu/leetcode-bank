```python
board = [[1,2,3],[4,0,5]]
# Output: 1
```

```python
from collections import deque
def slidingPuzzle(board):
    m, n = len(board), len(board[0])
    start = "".join(str(x) for row in board for x in row)
    target = "123450"
    distance = {start: 0}
    q = deque([start])
    
    while q:
        current = q.popleft()
        if current == target:
            return distance[current]
        for state in get_next_states(current, m, n):
            if state not in distance:
                distance[state] = distance[current] + 1
                q.append(state)
    return -1

def get_next_states(current, m, n):
    zero_index = current.index('0')
    x, y = zero_index // n, zero_index % n
    next_states = []
    
    for dx, dy in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
        nx, ny = x + dx, y + dy
        if nx < 0 or nx >= m or ny < 0 or ny >= n:
            continue
        new_state = list(current)
        new_state[zero_index], new_state[nx * n + ny] = new_state[nx * n + ny], new_state[zero_index]
        next_states.append("".join(new_state))
    return next_states  
        
```

```python
slidingPuzzle(board) # time O(N!) space O(N!)
```

```python
def slidingPuzzle_bi_directional(board):
    m, n = len(board), len(board[0])
    start = "".join(str(x) for row in board for x in row)
    target = "123450"
    if start == target:
        return 0

    step = 0
    forward_q = deque([start])
    backward_q = deque([target])
    forward_visited = set([start])
    backward_visited = set([target])
    
    while forward_q and backward_q:
        step += 1
        if extend_q(forward_q, forward_visited, backward_visited, m, n):
            return step

        step += 1
        if extend_q(backward_q, backward_visited, forward_visited, m, n):
            return step

    return -1

def extend_q(q, visited, other_visited, m, n):
    for _ in range(len(q)):
        current = q.popleft()
        for state in get_next_states(current, m, n):
            if state in visited:
                continue
            if state in other_visited:
                return True
            visited.add(state)
            q.append(state)
    return False    
        
    
```

```python
slidingPuzzle_bi_directional(board) # time O(sqrt(N!)) space O(sqrt(N!))
```
