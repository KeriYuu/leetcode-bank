```python
n = 4
m = 5
A = [(1,1),(0,1),(3,3),(3,4)]
# Output: [1,1,2,2]
```

```python
class UnionFind:
    def __init__(self):
        self.father = {}
        self.size = 0

    def add(self, var):
        if var not in self.father:
            self.father[var] = None
            self.size += 1

    def merge(self, var1, var2):
        root1 = self._find_father(var1)
        root2 = self._find_father(var2)
        if root1 != root2:
            self.father[root1] = root2
            self.size -= 1

    def _find_father(self, var):
        root = var
        while self.father[root] != None:
            root = self.father[root]

        while var != root:
            original_father = self.father[var]
            self.father[var] = root
            var = original_father
        return root
            

def numIslands2(n, m, A):
    uf = UnionFind()
    visited = [[False] * m for _ in range(n)]
    ans = []
    for x, y in A:
        # ~ O(1)
        if visited[x][y]:
            ans.append(uf.size)
            continue
        visited[x][y] = True
        uf.add((x, y))
        for dx, dy in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
            nx, ny = x + dx, y + dy
            if 0 <= nx < n and 0 <= ny < m and visited[nx][ny]:
                uf.merge((x, y), (nx, ny))
        ans.append(uf.size)
    return ans

numIslands2(n, m, A)
```
